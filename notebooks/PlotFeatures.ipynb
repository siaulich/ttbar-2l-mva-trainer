```python
def compute_closest_non_matching_jet(data_dict, data_config: DataConfig):
    padding_value = data_config.padding_value
    jet_eta = data_dict["jet_inputs"][:, :, 1]
    jet_phi = data_dict["jet_inputs"][:, :, 2]
    lep_eta = data_dict["lep_inputs"][:, :, 1]
    lep_phi = data_dict["lep_inputs"][:, :, 2]
    labels = data_dict["assignment"]
    compute_closest_non_matching_jet = np.full((jet_eta.shape[0], 2), -1, dtype=float)
    for i in range(jet_eta.shape[0]):
        for lep_idx in range(2):
            min_dR = float("inf")
            match_dR = -1
            for jet_idx in range(jet_eta.shape[1]):
                if jet_eta[i, jet_idx] == padding_value:
                    continue
                if labels[i, jet_idx, lep_idx]:
                    match_dR = np.sqrt(
                        (jet_eta[i, jet_idx] - lep_eta[i, lep_idx]) ** 2
                        + (jet_phi[i, jet_idx] - lep_phi[i, lep_idx]) ** 2
                    )
                    continue
                dR = np.sqrt(
                    (jet_eta[i, jet_idx] - lep_eta[i, lep_idx]) ** 2
                    + (jet_phi[i, jet_idx] - lep_phi[i, lep_idx]) ** 2
                )
                if dR < min_dR and not labels[i, jet_idx, lep_idx]:
                    min_dR = dR
            compute_closest_non_matching_jet[i, lep_idx] = min_dR - match_dR
    return np.min(compute_closest_non_matching_jet, axis=1)
```

```python
def compute_n_b_jets(data_dict, data_config: DataConfig):
    n_b_jets = np.sum(data_dict["jet_inputs"][:, :, 4] >= 2, axis=1)
    return n_b_jets
```

```python
from core.evaluation import c_han, c_hel, c_hel_test
from core.utils import lorentz_vector_from_PtEtaPhiE_array


def compute_c_hel(data_dict, data_config: DataConfig):
    top_PtEtaPhiE = data_dict["top_truth"][:, 0, :4]
    tbar_PtEtaPhiE = data_dict["top_truth"][:, 0, :4]
    lep_pos_PtEtaPhiE = data_dict["lepton_truth"][:, 0, :4]
    lep_neg_PtEtaPhiE = data_dict["lepton_truth"][:, 1, :4]
    c_hel_values = c_hel(
        lorentz_vector_from_PtEtaPhiE_array(top_PtEtaPhiE),
        lorentz_vector_from_PtEtaPhiE_array(tbar_PtEtaPhiE),
        (lep_pos_PtEtaPhiE),
        (lep_neg_PtEtaPhiE),
    )
    return c_hel_values


def compute_c_han(data_dict, data_config: DataConfig):
    top_PtEtaPhiE = data_dict["top_truth"][:, 0, :4]
    tbar_PtEtaPhiE = data_dict["top_truth"][:, 0, :4]
    lep_pos_PtEtaPhiE = data_dict["lepton_truth"][:, 0, :4]
    lep_neg_PtEtaPhiE = data_dict["lepton_truth"][:, 1, :4]
    c_han_values = c_han(
        lorentz_vector_from_PtEtaPhiE_array(top_PtEtaPhiE),
        lorentz_vector_from_PtEtaPhiE_array(tbar_PtEtaPhiE),
        (lep_pos_PtEtaPhiE),
        (lep_neg_PtEtaPhiE),
    )
    return c_han_values
```

```python
def fake_b_jet_rate(data_dict, data_config: DataConfig):
    assignment = data_dict["assignment"]
    b_tag_scores = data_dict["jet_inputs"][:, :, 4]
    fake_b_jets = (b_tag_scores >= 2) & (~assignment.any(axis=2))
    fake_b_jet_counts = np.sum(fake_b_jets, axis=1)
    return fake_b_jet_counts / (
        np.sum(b_tag_scores >= 2, axis=1) + 1e-6
    )  # Avoid division by zero


def fake_b_tag_rate(data_dict, data_config: DataConfig):
    assignment = data_dict["assignment"]
    b_tag_scores = data_dict["jet_inputs"][:, :, 4]
    fake_b_tags = (b_tag_scores >= 2) & (~assignment.any(axis=2))
    fake_b_tag_counts = np.sum(fake_b_tags, axis=1)
    N_jets = (b_tag_scores != data_config.padding_value).sum(axis=1)
    return (
        fake_b_tag_counts / (N_jets - 2 + 1e-6) * (N_jets > 2)
    )  # Avoid division by zero and negative rates


def b_jet_miss_rate(data_dict, data_config: DataConfig):
    assignment = data_dict["assignment"]
    b_tag_scores = data_dict["jet_inputs"][:, :, 4]
    missed_b_jets = (
        (assignment.any(axis=2))
        & (b_tag_scores < 2)
        & (b_tag_scores != data_config.padding_value)
    )
    missed_b_jet_counts = np.sum(missed_b_jets, axis=1)
    return missed_b_jet_counts / 2  # Assuming 2 true b-jets per event
```

```python
def truth_top_neutrino_pt(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 0, :3]
    neutrino_pt = (
        np.sqrt(neutrino_3vector[:, 0] ** 2 + neutrino_3vector[:, 1] ** 2) / 1e3
    )
    return neutrino_pt


def truth_tbar_neutrino_pt(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 1, :3]
    neutrino_pt = (
        np.sqrt(neutrino_3vector[:, 0] ** 2 + neutrino_3vector[:, 1] ** 2) / 1e3
    )
    return neutrino_pt


def truth_top_neutrino_eta(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 0, :3]
    neutrino_eta = np.arctanh(
        neutrino_3vector[:, 2]
        / np.sqrt(
            neutrino_3vector[:, 0] ** 2
            + neutrino_3vector[:, 1] ** 2
            + neutrino_3vector[:, 2] ** 2
        )
    )
    return neutrino_eta


def truth_tbar_neutrino_eta(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 1, :3]
    neutrino_eta = np.arctanh(
        neutrino_3vector[:, 2]
        / np.sqrt(
            neutrino_3vector[:, 0] ** 2
            + neutrino_3vector[:, 1] ** 2
            + neutrino_3vector[:, 2] ** 2
        )
    )
    return neutrino_eta


def truth_top_neutrino_phi(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 0, :3]
    neutrino_phi = np.arctan2(neutrino_3vector[:, 1], neutrino_3vector[:, 0])
    return neutrino_phi


def truth_tbar_neutrino_phi(data_dict, data_config: DataConfig):
    neutrino_3vector = data_dict["regression"][:, 1, :3]
    neutrino_phi = np.arctan2(neutrino_3vector[:, 1], neutrino_3vector[:, 0])
    return neutrino_phi
```

```python
def delta_R(jet_inputs, lepton_inputs):
    delta_eta = jet_inputs[1] - lepton_inputs[1]
    delta_phi = jet_inputs[2] - lepton_inputs[2]
    delta_phi = (delta_phi + np.pi) % (2 * np.pi) - np.pi
    return np.sqrt(delta_eta**2 + delta_phi**2)


def invariant_mass(jet_inputs, lepton_inputs):
    jet_e = jet_inputs[3]
    jet_px = jet_inputs[0] * np.cos(jet_inputs[2])
    jet_py = jet_inputs[0] * np.sin(jet_inputs[2])
    jet_pz = jet_inputs[0] * np.sinh(jet_inputs[1])

    lep_e = lepton_inputs[3]
    lep_px = lepton_inputs[0] * np.cos(lepton_inputs[2])
    lep_py = lepton_inputs[0] * np.sin(lepton_inputs[2])
    lep_pz = lepton_inputs[0] * np.sinh(lepton_inputs[1])

    total_e = jet_e + lep_e
    total_px = jet_px + lep_px
    total_py = jet_py + lep_py
    total_pz = jet_pz + lep_pz

    return np.sqrt(total_e**2 - total_px**2 - total_py**2 - total_pz**2)


data_plotter.reset_data_cuts()
fig, ax = data_plotter.plot_relational_jet_lepton_inputs(
    delta_R, r"$\Delta R(\ell, \text{jet})$", bins=50, range=(0, 5), region_tag="Inclusive"
)
ax.set_xlim((0, 5))

fig.savefig(PLOTS_DIR + "relational_jet_lepton_deltaR_distribution.pdf")


data_plotter.register_data_cut("non_training", "truth_ttbar_mass", lambda x: x < 390e3)
fig, ax = data_plotter.plot_relational_jet_lepton_inputs(
    delta_R, r"$\Delta R(\ell, \text{jet})$", bins=50, range=(0, 5), region_tag=r"$m_{t\bar{t}}<390 $GeV"
)
ax.set_xlim((0, 5))

fig.savefig(PLOTS_DIR + "relational_jet_lepton_deltaR_distribution_m_tt_390GeV.pdf")
data_plotter.reset_data_cuts()
```
